```python
# text clustering
# press releases
# each line represents a Press Release. Fields are as follows:
# id: Press release number (can be missing if included in text)
# title: Title of release
# contents: Text of release
# date: Posted date
# topics: Array of topic tags (if any provided)
# components Array of agencies & departments (if any provided)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import MiniBatchKMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
# upload the data
data = pd.read_json('combined.json', lines=True)
print("Loaded data shape:", data.shape)
data.head()
```

```python
# Tf-idf vectorization with improved parameters
tfidf = TfidfVectorizer(
    min_df=5,           # Bỏ qua từ xuất hiện trong ít hơn 5 văn bản
    max_df=0.95,        # Bỏ qua từ xuất hiện trong hơn 95% văn bản
    max_features=8000,  # Giữ tối đa 8000 từ phổ biến nhất
    stop_words='english',
    ngram_range=(1, 2)  # Thêm ngram range để nắm bắt cụm từ (1-2 từ)
)
tfidf.fit(data.contents)
text = tfidf.transform(data.contents)
print(f"Vectorized data shape: {text.shape}, Sparsity: {100.0 * (1 - text.nnz / (text.shape[0] * text.shape[1])):.2f}%")
```

```python
# Look for optimal number of clusters using elbow method
def find_optimal_clusters(data, max_k):
    iters = range(2, max_k+1, 2)
    
    sse = []
    for k in iters:
        kmeans = MiniBatchKMeans(
            n_clusters=k, 
            init_size=1024, 
            batch_size=2048, 
            random_state=42,
            n_init='auto'  # Auto parameter untuk phiên bản mới của scikit-learn
        )
        kmeans.fit(data)
        sse.append(kmeans.inertia_)
        print(f'Fit {k} clusters, SSE: {kmeans.inertia_:.2f}')
        
    f, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.plot(iters, sse, marker='o', linestyle='-', linewidth=2, markersize=8)
    ax.set_xlabel('Số lượng cụm (Clusters)', fontsize=12)
    ax.set_xticks(iters)
    ax.set_xticklabels(iters)
    ax.set_ylabel('SSE (Sum of Squared Errors)', fontsize=12)
    ax.set_title('Biểu đồ khuỷu tay (Elbow Plot) cho K-means Clustering', fontsize=14)
    ax.grid(True, linestyle='--', alpha=0.7)
    
    # Tính toán tỷ lệ giảm của SSE
    sse_ratio = [1.0]
    for i in range(1, len(sse)):
        sse_ratio.append(sse[i] / sse[i-1])
    
    # Tìm điểm gấp khúc của đường cong
    optimal_k = iters[np.argmin(sse_ratio) + 1] if len(sse_ratio) > 1 else iters[0]
    print(f"Số lượng cụm được đề xuất: {optimal_k}")
    
    # Vẽ đường dọc tại số lượng cụm được đề xuất
    ax.axvline(x=optimal_k, color='r', linestyle='--', label=f'Đề xuất: k={optimal_k}')
    ax.legend()
    
    return optimal_k
```

```python

# Tìm số lượng cụm tối ưu
optimal_k = find_optimal_clusters(text, 20)

# Set k to optimal value (hoặc sử dụng 6 như ví dụ gốc)
n_clusters = optimal_k  # Có thể thay đổi thành optimal_k
clusters = MiniBatchKMeans(
    n_clusters=n_clusters, 
    init_size=1024, 
    batch_size=2048, 
    random_state=42,
    n_init='auto'  # Required parameter in newer scikit-learn versions
).fit_predict(text)
```

```python
# Trực quan hóa kết quả phân cụm với PCA và t-SNE
def plot_tsne_pca(data, labels, n_clusters=None):
    max_label = max(labels)
    if n_clusters is None:
        n_clusters = max_label + 1  # Số lượng cụm bằng nhãn lớn nhất + 1 nếu không được chỉ định
    
    max_items = np.random.choice(range(data.shape[0]), size=3000, replace=False)
    
    # Sửa lỗi np.matrix với np.asarray
    pca = PCA(n_components=2).fit_transform(np.asarray(data[max_items,:].todense()))
    tsne = TSNE(
        n_components=2,
        perplexity=30,
        learning_rate='auto',
        init='pca',
        random_state=42
    ).fit_transform(PCA(n_components=50).fit_transform(np.asarray(data[max_items,:].todense())))
    
    idx = np.random.choice(range(pca.shape[0]), size=300, replace=False)
    label_subset = labels[max_items]
    label_subset = [cm.tab10(i/max_label) for i in label_subset[idx]]  # Sử dụng tab10 colormap thay vì hsv
    
    f, ax = plt.subplots(1, 2, figsize=(16, 7))
    
    # PCA plot
    ax[0].scatter(pca[idx, 0], pca[idx, 1], c=label_subset, s=50, alpha=0.8)
    ax[0].set_title('PCA Cluster Plot', fontsize=14)
    ax[0].set_xlabel('Principal Component 1', fontsize=12)
    ax[0].set_ylabel('Principal Component 2', fontsize=12)
    ax[0].grid(True, linestyle='--', alpha=0.3)
    
    # t-SNE plot
    ax[1].scatter(tsne[idx, 0], tsne[idx, 1], c=label_subset, s=50, alpha=0.8)
    ax[1].set_title('t-SNE Cluster Plot', fontsize=14)
    ax[1].set_xlabel('t-SNE Dimension 1', fontsize=12)
    ax[1].set_ylabel('t-SNE Dimension 2', fontsize=12)
    ax[1].grid(True, linestyle='--', alpha=0.3)
    
    plt.tight_layout()
    
    # Thêm chú thích cho các cụm
    handles = []
    for i in range(n_clusters):
        handles.append(plt.Line2D([0], [0], marker='o', color='w', 
                                 markerfacecolor=cm.tab10(i/max_label), markersize=10, 
                                 label=f'Cluster {i}'))
    
    # Thêm legend vào dưới hình
    plt.figlegend(handles=handles, loc='lower center', ncol=n_clusters, 
                 bbox_to_anchor=(0.5, -0.05), frameon=True)
    
    plt.subplots_adjust(bottom=0.15)
    
# Vẽ biểu đồ
plot_tsne_pca(text, clusters, n_clusters)
```

```python
# Hiển thị từ khóa hàng đầu dựa trên điểm TF-IDF để xem xu hướng
def get_top_keywords(data, clusters, labels, n_terms):
    # Chuyển đổi ma trận thưa thành DataFrame và tính trung bình theo cụm
    df = pd.DataFrame(np.asarray(data.todense())).groupby(clusters).mean()
    
    # Dictionary để lưu trữ từ khóa hàng đầu cho mỗi cụm
    cluster_keywords = {}
    
    # Với mỗi cụm, lấy n_terms từ khóa hàng đầu
    for i, r in df.iterrows():
        # Sắp xếp các chỉ số theo giá trị TF-IDF giảm dần
        top_indices = np.argsort(r)[-n_terms:][::-1]
        top_terms = [labels[t] for t in top_indices]
        
        print(f'\nCluster {i}:')
        print(', '.join(top_terms))
        
        # Lưu từ khóa cho mỗi cụm
        cluster_keywords[i] = top_terms
    
    # Tính toán số hàng và số cột cho subplots
    num_clusters = len(cluster_keywords)
    if num_clusters <= 3:
        rows, cols = 1, num_clusters
    else:
        rows = (num_clusters + 2) // 3  # Làm tròn lên
        cols = min(3, num_clusters)
    
    # Tạo biểu đồ trực quan hóa từ khóa hàng đầu
    plt.figure(figsize=(14, 4*rows))
    
    for idx, (i, terms) in enumerate(cluster_keywords.items()):
        plt.subplot(rows, cols, idx+1)  # Sử dụng idx thay vì i
        y_pos = np.arange(len(terms[:10]))  # Top 10 terms
        values = [df.iloc[i][np.where(labels == term)[0][0]] for term in terms[:10]]
        
        bars = plt.barh(y_pos, values, align='center', alpha=0.8, color=cm.tab10(i/max(cluster_keywords.keys())))
        plt.yticks(y_pos, terms[:10])
        plt.title(f'Cluster {i} Top Terms', fontsize=12)
        plt.xlabel('TF-IDF Score', fontsize=10)
        
        # Thêm giá trị vào các thanh
        for bar in bars:
            width = bar.get_width()
            plt.text(width + 0.01, bar.get_y() + bar.get_height()/2, 
                    f'{width:.2f}', ha='left', va='center', fontsize=8)
    
    plt.tight_layout()
    plt.suptitle('Top Keywords by Cluster', fontsize=16, y=1.02)
    
    return cluster_keywords
    
# Hiển thị từ khóa hàng đầu cho mỗi cụm
# Sử dụng get_feature_names_out() thay vì get_feature_names()
keywords = get_top_keywords(text, clusters, tfidf.get_feature_names_out(), 15)
```

```python
# Phân tích phân phối cụm
cluster_counts = pd.Series(clusters).value_counts().sort_index()
print("\nCluster distribution:")
for i, count in cluster_counts.items():
    percentage = 100 * count / len(clusters)
    print(f"Cluster {i}: {count} documents ({percentage:.2f}%)")

# Vẽ biểu đồ phân phối cụm
plt.figure(figsize=(10, 6))
bars = plt.bar(cluster_counts.index, cluster_counts.values, color=[cm.tab10(i/max(clusters)) for i in cluster_counts.index])
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Number of Documents', fontsize=12)
plt.title('Document Distribution Across Clusters', fontsize=14)
plt.xticks(cluster_counts.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Thêm số lượng và phần trăm lên các thanh
for bar in bars:
    height = bar.get_height()
    percentage = 100 * height / len(clusters)
    plt.text(bar.get_x() + bar.get_width()/2, height + 5, 
             f'{int(height)}\n({percentage:.1f}%)', 
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
```
